```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["deck", "alive", "class", "who", "adult_male", "embark_town"], axis=1, inplace=True)
```

```python
df.info()
```

```python
df["age"].fillna(df["age"].mean(), inplace=True)
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)
df.info()
```

```python
# Import LabelEncoder from sklearn
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
```

```python
df=df.astype(int)
```

```python
df.head()
```

```python
# Separate features (X) and target (y)

# Drop 'survived' column from dataset → this becomes input features
X = df.drop("survived", axis=1)

# Select 'survived' column → this is the target variable
y = df["survived"]


# Import function to split dataset
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% data for testing, 80% for training
    random_state=42     # ensures same split every time (reproducibility)
)
```

```python
# Import Logistic Regression model
from sklearn.linear_model import LogisticRegression

# Create model object
model = LogisticRegression()

# Train (fit) the model using training data
model.fit(X_train, y_train)
```

```python
# Make predictions on test data
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Import evaluation metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

```

```python
# Calculate accuracy (how many predictions are correct)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

```

```python
# Confusion Matrix (actual vs predicted)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python

# Detailed report (precision, recall, f1-score)
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

# Create a KNN model object (e.g., with n_neighbors=5)
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train the KNN model using the scaled training data
knn_model.fit(X_train_scaled, y_train)
```

```python
# Make predictions on the scaled test data
y_pred_knn = knn_model.predict(X_test_scaled)
```

### KNN Model Evaluation

```python
# Calculate accuracy for KNN
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy:", accuracy_knn)

# Confusion Matrix for KNN
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("\nKNN Confusion Matrix:\n", cm_knn)

# Classification Report for KNN
report_knn = classification_report(y_test, y_pred_knn)
print("\nKNN Classification Report:\n", report_knn)
```

```python
# Import Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Naive Bayes model object
nb_model = GaussianNB()

# Train the Naive Bayes model using the scaled training data
nb_model.fit(X_train, y_train)

# Make predictions on the scaled test data
y_pred_nb = nb_model.predict(X_test)

# --- Naive Bayes Model Evaluation ---

# Calculate accuracy for Naive Bayes
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Naive Bayes Accuracy:", accuracy_nb)

# Confusion Matrix for Naive Bayes
cm_nb = confusion_matrix(y_test, y_pred_nb)
print("\nNaive Bayes Confusion Matrix:\n", cm_nb)

# Classification Report for Naive Bayes
report_nb = classification_report(y_test, y_pred_nb)
print("\nNaive Bayes Classification Report:\n", report_nb)
```

#SVM

```python
# Import SVM model
from sklearn.svm import SVC

# Import evaluation metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize SVM model
# kernel='rbf' is default (works well in most cases)
model_SVM = SVC(kernel='rbf', random_state=42)

# Train the model
model_SVM.fit(X_train_scaled, y_train)

# Make predictions
y_pred_SVM = model_SVM.predict(X_test_scaled)

# Display predictions
y_pred_SVM

# Calculate accuracy
acc_SVM = accuracy_score(y_test, y_pred_SVM)
print("SVM Accuracy:", acc_SVM)

# Confusion Matrix
cm_SVM = confusion_matrix(y_test, y_pred_SVM)
print("\nSVM Confusion Matrix:\n", cm_SVM)

# Classification Report
report_SVM = classification_report(y_test, y_pred_SVM)
print("\nSVM Classification Report:\n", report_SVM)
```
